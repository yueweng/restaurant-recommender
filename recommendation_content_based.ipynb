```python
import pandas as pd

restaurants_df = pd.read_csv('data/restaurants.csv')
filtered_restaurants_df = restaurants_df.drop(columns=['Unnamed: 0', '_id', 'image_url', 'url', 'phone', 'address', 'kids', 'city', 'state', 'zipcode', 'description'])
```

```python
str_cols = ['ambience', 'noise_level', 'alcohol', 'meal']

for col in str_cols:
    filtered_restaurants_df[col].fillna("", inplace=True)
filtered_restaurants_df['price'].fillna(0, inplace=True)

cols = ['health_score', 'reservations', 'take_out', 'apple_pay', 'bike_parking', 'happy_hour',
        'wifi', 'dogs', 'drive_thru', 'vegan_options', 'delivery', 'credit_cards', 'google_pay', 
       'parking', 'outdoor_seating', 'tv', 'waiter_service', 'caters', 'smoking', 'dancing', 
       'wheelchair', 'coat', 'groups', 'happy_hour_sp']

for col in cols:
    filtered_restaurants_df[col].fillna(0, inplace=True)

filtered_restaurants_df.info()
```

```python
s = filtered_restaurants_df['ambience'].str.split(", ")
ambience_dummies = pd.get_dummies(s.apply(pd.Series).stack()).sum(level=0)
ambience_dummies
```

```python
cui = filtered_restaurants_df['cuisines'].str.split(", ")
cuisines_dummies = pd.get_dummies(cui.apply(pd.Series).stack()).sum(level=0)
cuisines_dummies
```

```python
meal = filtered_restaurants_df['meal'].str.split(", ")
meal_dummies = pd.get_dummies(meal.apply(pd.Series).stack()).sum(level=0)
meal_dummies
```

```python
parking = filtered_restaurants_df['parking'].str.split(", ")
parking_dummies = pd.get_dummies(parking.apply(pd.Series).stack()).sum(level=0)
parking_dummies.rename(columns={'Yes': 'Parking_yes'}, inplace=True)
parking_dummies
```

```python
reviews_df = pd.read_csv('data/reviews.csv', encoding='utf-8', engine='python')
reviews_df
```

```python
ratings_df = pd.DataFrame({'userid': reviews_df['userid'], 'restaurant_id': reviews_df['restaurant_id'], 'stars': reviews_df['stars']})
ratings_df
```

```python
restaurants_name_df = pd.DataFrame({'id': restaurants_df['_id'], 'title': restaurants_df['title']})
restaurants_name_df
```

```python
users_df = pd.read_csv('data/users.csv')
users_df
```

```python
users_name_df = pd.DataFrame({'userid': users_df['userid'], 'name': users_df['name']})
users_name_df
```

```python
combine_df = ratings_df.merge(restaurants_name_df, left_on='restaurant_id', right_on='id')
combine_df = combine_df.merge(users_name_df, left_on='userid', right_on='userid')
combine_df.drop(columns=['id'], inplace=True)
```

```python
ratings_matrix = combine_df.pivot_table(index='userid', columns='title', values='stars').fillna(0)
```

```python
ratings_matrix
```

```python
sub_ratings_matrix = ratings_matrix[:30000]
sub_ratings_matrix
```

```python
from scipy.sparse.linalg import svds

# print(sub_ratings_matrix.shape[1])

u, s, vt = svds(sub_ratings_matrix, 10)
```

```python
import numpy as np

sigma = np.diag(s)
```

```python
predictions = u @ sigma @ vt
index = predictions[0].argsort()[-10:][::-1]


for ind in index:
    print(sub_ratings_matrix.columns.to_list()[ind])
```
