```python
import pandas as pd

restaurants_df = pd.read_csv('data/restaurants.csv')
filtered_restaurants_df = restaurants_df.drop(columns=['Unnamed: 0', '_id', 'image_url', 'url', 'phone', 'address', 'kids', 'city', 'state', 'zipcode', 'description'])
```

```python
%run 'src/recommendations.py'
```

```python
str_cols = ['ambience', 'noise_level', 'alcohol', 'meal']

for col in str_cols:
    filtered_restaurants_df[col].fillna("", inplace=True)
filtered_restaurants_df['price'].fillna(0, inplace=True)

cols = ['health_score', 'reservations', 'take_out', 'apple_pay', 'bike_parking', 'happy_hour',
        'wifi', 'dogs', 'drive_thru', 'vegan_options', 'delivery', 'credit_cards', 'google_pay', 
       'parking', 'outdoor_seating', 'tv', 'waiter_service', 'caters', 'smoking', 'dancing', 
       'wheelchair', 'coat', 'groups', 'happy_hour_sp']

for col in cols:
    filtered_restaurants_df[col].fillna(0, inplace=True)

filtered_restaurants_df.info()
```

```python
filtered_restaurants_df
```

```python
s = filtered_restaurants_df['ambience'].str.split(", ")
ambience_dummies = pd.get_dummies(s.apply(pd.Series).stack()).sum(level=0)
ambience_dummies
```

```python
cui = filtered_restaurants_df['cuisines'].str.split(", ")
cuisines_dummies = pd.get_dummies(cui.apply(pd.Series).stack()).sum(level=0)
cuisines_dummies
```

```python
meal = filtered_restaurants_df['meal'].str.split(", ")
meal_dummies = pd.get_dummies(meal.apply(pd.Series).stack()).sum(level=0)
meal_dummies
```

```python
parking = filtered_restaurants_df['parking'].str.split(", ")
parking_dummies = pd.get_dummies(parking.apply(pd.Series).stack()).sum(level=0)
parking_dummies.rename(columns={'Yes': 'Parking_yes'}, inplace=True)
parking_dummies
```

```python
dum_cols = ['neighborhood', 'noise_level', 'alcohol']

dum = {}
for col in dum_cols:
    dum[col] = pd.get_dummies(filtered_restaurants_df[col])
```

```python
concat_rest_df = pd.concat([filtered_restaurants_df, ambience_dummies, cuisines_dummies, meal_dummies, dum['alcohol'], dum['noise_level']], axis=1)
concat_rest_df
```

```python
concat_rest_df.drop(columns=['title','neighborhood', 'cuisines', 'ambience', 'meal', 'alcohol', 'noise_level'], inplace=True)
cols = ['reservations', 'take_out', 'apple_pay', 'bike_parking', 'happy_hour',
        'wifi', 'dogs', 'drive_thru', 'vegan_options', 'delivery', 'credit_cards', 'google_pay', 
       'parking', 'outdoor_seating', 'tv', 'waiter_service', 'caters', 'smoking', 'dancing', 
       'wheelchair', 'coat', 'groups', 'happy_hour_sp']
for col in cols:
    concat_rest_df.drop(columns=[col], inplace=True)
```

```python
concat_rest_df.fillna(0, inplace=True)
concat_rest_df.drop(columns=['reviewCount'], inplace=True)
concat_rest_df
```

```python
condensed_df = pd.DataFrame({'price': concat_rest_df['price'], 'stars': concat_rest_df['stars']})
# con_df = pd.concat([condensed_df, cuisines_dummies], axis=1)
# con_df

example_df = pd.concat([cuisines_dummies, condensed_df], axis=1)
example_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cos_sim = cosine_similarity(example_df, example_df)
cos_sim
```

```python
cos_sim.shape
```

```python
# Try with Marufuku Ramen SF

# for idx, titles in filtered_restaurants_df.items():
#     for title in titles:
title = 'Marufuku Ramen SF'
print("Restaurant: {}".format(title))

index = filtered_restaurants_df[filtered_restaurants_df['title'] == title].index
compare_sim = cos_sim[index]
# print(compare_sim)

sim = {}
# iterate each row of cosine similarity matrix and compute the distance
for idx, row in enumerate(cos_sim):
#     print(row)
    if index != idx:
        sim[idx] = cosine_similarity(compare_sim.reshape(1, -1), row.reshape(1, -1))[0][0].astype(float)

sort_values = sorted(sim.items(), key=lambda kv: kv[1], reverse=True)
top_n = sort_values[:20]

print("Recommendations")
for row, value in top_n:
    print(value)
    print(filtered_restaurants_df.iloc[row]['title'])

print("-------------------")
```

```python
from scipy.sparse.linalg import svds

# print(sub_ratings_matrix.shape[1])

u, s, vt = svds(ratings_matrix, 10)
```

```python
import numpy as np

sigma = np.diag(s)
```

```python
benchmark = []
# Iterate over all algorithms
for algorithm in [SVD()
                  , NMF()
                  , NormalPredictor()
                  , CoClustering()
                  , BaselineOnly()
                 ]:
    # Perform cross validation
    results = cross_validate(algorithm
                             , data
                             , measures=['RMSE']
                             , cv=5
                             , verbose=True)
    
    # Get results & append algorithm name
    tmp = pd.DataFrame.from_dict(results).mean(axis=0)
    tmp = tmp.append(pd.Series([str(algorithm).split(' ')[0].split('.')[-1]], index=['Algorithm']))
    benchmark.append(tmp)
    
pd.DataFrame(benchmark).set_index('Algorithm').sort_values('test_rmse')
```
