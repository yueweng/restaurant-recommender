```python
from pymongo import MongoClient
from bs4 import BeautifulSoup
import requests
import pandas as pd

client = MongoClient('localhost', 27017)
yelp = client['yelp']
restaurants = yelp['restaurants']
processed = yelp['processed']
users = yelp['users']
```

```python
restaurants_df = pd.DataFrame(list(restaurants.find()))
restaurants_df
```

```python
restaurants_df.to_csv('data/restaurants.csv', ",")
```

```python
restaurants_df[restaurants_df.duplicated(['title'])]
```

```python
restaurants_condensed = pd.DataFrame({'title': restaurants_df['title'], 'reviewCount': restaurants_df['reviewCount']})
restaurants_condensed.sort_values(by=['reviewCount'], ascending=False, inplace=True)
restaurants_condensed
```

```python
user_reviews_df = pd.DataFrame(list(users.find()))
user_reviews_df
```

```python
from bson import ObjectId
user_reviews_df[user_reviews_df['restaurant_id'] == ObjectId('5e500cd33296483cf9246ec3')]
```

```python
user_reviews_df[user_reviews_df.duplicated(['review', 'name'])]
```

```python
user_att = ['userid', 'name', 'url', 'city', 'state', 'friends', 'reviews', 'photos']
user = {}
for att in user_att:
    user[att] = user_reviews_df[att]


users_df = pd.DataFrame(user)
users_df
```

```python
users_df[users_df.duplicated(['userid'])]
```

```python
users_df.drop_duplicates(subset='userid', keep='first', inplace=True)
```

```python
users_df
```

```python
reviews_att = ['userid', 'restaurant_id', 'review', 'date', 'stars', 'Useful', 'Funny', 'Cool']
reviews = {}
for att in reviews_att:
    reviews[att] = user_reviews_df[att]


reviews_df = pd.DataFrame(reviews)
```
