```python
import pandas as pd

restaurants_df = pd.read_csv('data/restaurants.csv')
restaurants_df.drop(columns=['Unnamed: 0'], inplace=True)
restaurants_df
```

```python
import matplotlib.pyplot as plt

def neighborhood_dynamics(restaurants_df):
    neighborhood_group = restaurants_df.groupby('neighborhood').agg({'price': 'mean', 'stars': 'mean', 'reviewCount': 'sum', '_id': 'count'}).reset_index()
    neighborhood_group.sort_values(by=['reviewCount'], ascending=False, inplace=True)
    plot_neighborhood_popularity(neighborhood_group)
    

def plot_neighborhood_popularity(neighborhood_group, n=25):
    top_n = neighborhood_group[:n]
    fig, ax = plt.subplots(figsize=[20, 10])
    
    top_n.sort_values('reviewCount', inplace=True)
    
    ax.barh(top_n['neighborhood'], top_n['reviewCount'])
    ax.set_title('Popularity of Restaurants broken down by Neighborhood', fontsize=20, pad=20)
    ax.set_xlabel('Number of Reviews', fontsize=20)
    ax.set_ylabel('Neighborhood', fontsize=20)
    
    plt.tick_params(axis='x', which='major', labelsize=18)
    plt.tick_params(axis='y', which='major', labelsize=20)

    plt.tight_layout()
    


neighborhood_dynamics(restaurants_df)
```

```python
# ratings vs popularity
# plot all restaurants with x as the number of stars the restaurant has 
# and y will be the total number of reviews (popularity) in that category
# can also include the number of restaurants for each category, average of reviews per category
def restaurants_ratings(restaurants_df):
    reviews_group = restaurants_df.groupby('stars').agg({'_id': 'count', 'reviewCount': 'sum'}).reset_index().rename(columns={'_id': 'count'})
    fig, ax = plt.subplots(figsize=[20, 10])

    ax.bar(reviews_group['stars'], reviews_group['count'], width=0.4)
    ax.set_title('Num of Restaurants vs Ratings', fontsize=20, pad=20)
    ax.set_xlabel('Ratings', fontsize=20)
    ax.set_ylabel('Num of Restaurants', fontsize=20)

    plt.tick_params(axis='x', which='major', labelsize=18)
    plt.tick_params(axis='y', which='major', labelsize=15)

restaurants_ratings(restaurants_df)
```

```python
def ambience_df(restaurants_df):
    s = restaurants_df['ambience'].str.split(", ")
    ambience_dummies = pd.get_dummies(s.apply(pd.Series).stack()).sum(level=0)
    ambience_cols = ambience_dummies.columns
    rest_att = ['_id', 'title', 'price', 'stars', 'reviewCount']
    rest = {}
    for att in rest_att:
        rest[att] = restaurants_df[att]


    rest_condensed_df = pd.DataFrame(rest)
    ambience_df = pd.concat([rest_condensed_df, ambience_dummies], axis=1)
    amb = {}
    for col in ambience_cols:
        column = col.strip()
        for k, v in ambience_df[column].value_counts().items():
            if k == 1.0:
                amb[column] = v
    amb_df = pd.DataFrame(list(amb.items()), columns=['Ambience', 'Count'])
    amb_df.sort_values(by=['Count'], ascending=False, inplace=True)
    plot_ambience_overview(amb_df)
    
def plot_ambience_overview(amb_df):
    fig, ax = plt.subplots(figsize=[20, 10])
    ax.bar(amb_df['Ambience'], amb_df['Count'])

    ax.set_title('Ambience Overview', fontsize=20, pad=20)
    ax.set_xlabel('Ambience', fontsize=20)
    ax.set_ylabel('Num of Restaurants', fontsize=20)

    plt.tick_params(axis='x', which='major', labelsize=18)
    plt.tick_params(axis='y', which='major', labelsize=15)
    
ambience_df(restaurants_df)
```

```python
alcohol_dummies = pd.get_dummies(restaurants_df['alcohol'])
alcohol_cols = alcohol_dummies.columns

alcohol_df = pd.concat([rest_condensed_df, alcohol_dummies], axis=1)

al = {}
for col in alcohol_cols:
    for k, v in alcohol_df[col].value_counts().items():
        if k == 1.0:
            al[col] = v

al_df = pd.DataFrame(list(al.items()), columns=['Alcohol', 'Count'])
al_df.sort_values(by=['Alcohol'], inplace=True)

fig, ax = plt.subplots(figsize=[20, 10])
ax.bar(al_df['Alcohol'], al_df['Count'])
```

```python
def get_alcohol_popularity(restaurants_df):
    alcohol_dummies = pd.get_dummies(restaurants_df['alcohol'])
    alcohol_cols = alcohol_dummies.columns
    aldf = {}
    for col in alcohol_cols:
        al_rev = alcohol_df.groupby(col).agg({'reviewCount': 'sum'}).reset_index()
        aldf[col] = al_rev[al_rev[col] == 1.0]['reviewCount'][1]

    al_df = pd.DataFrame(list(aldf.items()), columns=['Alcohol', 'Popularity'])
    al_df.sort_values(by=['Alcohol'], inplace=True)
    return al_df

def plot_alcohol_popularity_graph(al_df):
    fig, ax = plt.subplots(figsize=[20, 10])
    ax.bar(al_df['Alcohol'], al_df['Popularity'])
    
def alcohol_overview(restaurants_df):
    al_df = get_alcohol_popularity(restaurants_df)
    plot_alcohol_popularity_graph(al_df)
    
alcohol_overview(restaurants_df)
```

```python
users_df = pd.read_csv('data/users.csv')
users_df
```

```python
sf_users = users_df[users_df['city'] == 'San Francisco']
len(sf_users)/len(users_df)
```

```python
ca_states = ['San Francisco', 'Los Angeles', 'Berkeley', 'Sacremento', 'San Jose', 'San Mateo', 'Menlo Park', 'South San Francisco']
for idx, user in users_df.iterrows():
    for state in ca_states:
        if user['state'] == state:
            users_df.loc[idx, 'state'] = 'CA'
            users_df.loc[idx, 'city'] = state
```

```python
users_group = users_df.groupby('state').agg({'userid': 'count'}).reset_index().rename(columns={'userid': 'count'})
users_group.sort_values(by=['count'], ascending=False, inplace=True)
top_users_state = users_group[:15]
top_users_state
```

```python
fig, ax = plt.subplots(figsize=[20, 10])

ax.bar(top_users_state['state'], top_users_state['count'])
```

```python
users_df.drop_duplicates(subset='userid', keep='first', inplace=True)
reviews_group = users_df.sort_values(by='reviews', ascending=False)
reviews_group[reviews_group['state'] == 'CA']
```

```python
city_review = reviews_group.groupby('city').agg({'state': 'first','userid': 'count'}).reset_index()
top_cities_reviews = city_review.sort_values(by='userid', ascending=False)[:25]
top_cities_reviews
```

```python
import random
unique_states = top_cities_reviews['state'].unique()

st = {}
for state in unique_states:
    r = random.uniform(0, 1)
    g = random.uniform(0, 1)
    b = random.uniform(0, 1)
    st[state] = [r, g, b]

st
```

```python
fig, ax = plt.subplots(figsize=[20, 10])

top_cities_reviews.sort_values('userid', inplace=True)

for idx, city in top_cities_reviews.iterrows():
    ax.barh(city['city'], city['userid'], color=st[city['state']], label=city['state'])
    
hand, labl = ax.get_legend_handles_labels()
lablout = []
handout = []
for h,l in zip(hand,labl):
    if l not in lablout:
        lablout.append(l)
        handout.append(h)

ax.set_title('Reviewers by City', fontsize=20, pad=20)
ax.set_xlabel('Num of Users', fontsize=20)
ax.set_ylabel('Cities', fontsize=20)

plt.tick_params(axis='x', which='major', labelsize=18)
plt.tick_params(axis='y', which='major', labelsize=15)
plt.legend(handout, lablout, fontsize=20)
```
