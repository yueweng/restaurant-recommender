```python
ratings_matrix = combine_df.pivot_table(index=group_users, columns='title', values='stars').fillna(0)
ratings_matrix
```

```python
import numpy as np
rm = ratings_matrix.as_matrix()
user_ratings_mean = np.mean(rm, axis = 1)
rm_demeaned = rm - user_ratings_mean.reshape(-1, 1)
rm_demeaned
```

```python
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(rm_demeaned, k = 10)
sigma = np.diag(sigma)
```

```python
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_ratings_mean.reshape(-1, 1)
preds_df = pd.DataFrame(all_user_predicted_ratings, columns = ratings_matrix.columns)
preds_df
```

```python
benchmark = []
# Iterate over all algorithms
for algorithm in [SVD()
                  , NMF()
                  , NormalPredictor()
                  , CoClustering()
                  , BaselineOnly()
                 ]:
    # Perform cross validation
    results = cross_validate(algorithm
                             , data
                             , measures=['RMSE']
                             , cv=5
                             , verbose=True)
    
    # Get results & append algorithm name
    tmp = pd.DataFrame.from_dict(results).mean(axis=0)
    tmp = tmp.append(pd.Series([str(algorithm).split(' ')[0].split('.')[-1]], index=['Algorithm']))
    benchmark.append(tmp)
    
pd.DataFrame(benchmark).set_index('Algorithm').sort_values('test_rmse')
```
